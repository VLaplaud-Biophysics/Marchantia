## Define analysis functions

###  Change reference from time to area

Adds an 'AreaGroup' column in dataframe, similar in function to 'Img', to regroupe gemmae of similar area instaed of gemmae from the same timepoint. 


```python
def time2area(GD,CD,StackList,nbin):
    
    warnings.filterwarnings("ignore")  
    
    # Compute bins over all areas
    n,bins = np.histogram(GD['Area'],bins=nbin)
       
    # Adds the area group for global dataframe
    GD['AreaGroup'] = pd.cut(GD['Area'],bins,labels=np.arange(1,nbin+1)).to_numpy()
    
    for s in StackList:
   
        nimg = int(1+np.max(CD.loc[s, 'Img']))
        
        for i in range(nimg):

            CD.loc[(CD.index==s) &(CD['Img']==i),'AreaGroup'] = GD.loc[
                (GD.index==s) &(GD['Img']==i),'AreaGroup']
            
    warnings.filterwarnings("default")  
    
    return(GD,CD)

print('Function defined.')
```

###  Mean contour over all gemma


####  In time

Computation only for the number of images of the shortest video. Ponctually one gemma can be ignored for one time point if the focus on the video is bad.


#####  As mean (X,Y) position based on parametrisation

```python
def computeMeanContourTime(StackList,CD,GD):
    
    # Number of image in the shortest videos
    nmin = 1 + int(np.min([np.max(CD.loc[s, 'Img']) for s in StackList]))
    
    npts = len(CD[CD['Img']==0].loc[StackList[0],'Xparam'])
    
    # Storing variables
    alignedXs = np.empty((npts,len(StackList)))
    alignedXs[:] = np.nan
    alignedYs = np.empty((npts,len(StackList)))
    alignedYs[:] = np.nan
    
    for i in range(nmin):
        
        print('Computing mean contour for image : ' + str(i) + '/' + str(nmin).ljust(5), end='\r')
        
        # Not normalized
        
        # On each image, retrieve aligned contour for each gemma
        
        for ii,s in zip(range(len(StackList)),StackList):
           
            Xn = CD.loc[(CD.index == s) & (CD['Img'] == i),'Xparam'] 
            Yn = CD.loc[(CD.index == s) & (CD['Img'] == i),'Yparam']            
            
            if bool(GD.loc[(GD.index == s) &
                                   (GD['Img'] == i),'GoodFocus'].to_numpy) == True:
                alignedXs[:,ii] = Xn
                alignedYs[:,ii] = Yn

        # Computing mean of aligned contours        
        meanX = np.nanmean(alignedXs,1)
        meanY = np.nanmean(alignedYs,1)
        
        # Storing contour data
        data = {'Xparam':meanX,
                    'Yparam':meanY,
                    'Img':i*np.ones(len(meanX))} 
      
        newdata = pd.DataFrame(data=data,index = np.repeat('meanContTime',len(meanX)))
        
        CD = CD.append(newdata) # adding to global dataframe
        
        # Normalized
        
        # Storing variables
        alignedXNs = np.empty((npts,len(StackList)))
        alignedXNs[:] = np.nan
        alignedYNs = np.empty((npts,len(StackList)))
        alignedYNs[:] = np.nan
        
        # On each image, retrieve aligned contour for each gemma
        
        for ii,s in zip(range(len(StackList)),StackList):
           
            XnN = CD.loc[(CD.index == s) & (CD['Img'] == i),'XparamNorm']  
            YnN = CD.loc[(CD.index == s) & (CD['Img'] == i),'YparamNorm']            
            
            if bool(GD.loc[(GD.index == s) &
                                   (GD['Img'] == i),'GoodFocus'].to_numpy) == True:
                alignedXNs[:,ii] = XnN
                alignedYNs[:,ii] = YnN

        # Computing mean of aligned contours        
        meanXN = np.nanmean(alignedXNs,1)
        meanYN = np.nanmean(alignedYNs,1)
        
        # Storing contour data
        data = {'Xparam':meanXN,
                    'Yparam':meanYN,
                    'Img':i*np.ones(len(meanXN))} 
      
        newdata = pd.DataFrame(data=data,index = np.repeat('normMeanContTime',len(meanX)))
        
        CD = CD.append(newdata) # adding to global dataframe

    print('Computing mean contour for image : ' + str(nmin) + '/' + str(nmin).ljust(5), end='\n\n')
        
    return(CD)

print('Function defined.')
```

##### As a probability density of being inside a contour

###### Get contour from binary

```python
def ContourFromBinImg(BinImg,Scale):   
    # Scale in µm/px
    
    # Detecting edges and computing their position
    edge = sobel(BinImg)>threshold_otsu(sobel(BinImg))
    indices = np.where(edge != [0])
    
    # Computing center of gemma using euclidian distance transform
    dist = cv.distanceTransform(BinImg, cv.DIST_L2, 3)
    regions = regionprops(BinImg,dist)
    center = regions[0].weighted_centroid
    
    # Edge coorinate relative to the center
    relativeIndicesX = (indices[0]-center[0])*Scale
    relativeIndicesY = (indices[1]-center[1])*Scale
    
    # Edge coordinate in circular system
    Theta,Radius = vf.ToCirc(relativeIndicesX,relativeIndicesY, angle = 'deg')

    # interpolation of contour on regular grid
    extraR = Radius[np.argmin(np.abs(Theta-180))]   
    f = interp1d(Theta, Radius, fill_value=(extraR,extraR),bounds_error=False)
    Thetanew = np.linspace(-179.5,180,720,endpoint=True)
    Radiusnew = f(Thetanew)  

    # Smoothing
    SmoothRadius = savgol_filter(Radiusnew,7,3)  
    
    # Recompute X and Y from inerpolated radius
    X,Y = vf.ToCart(Thetanew,SmoothRadius,angle = 'deg')
    
    # Sorting coordinate based on angle
    FullCoordinates = np.stack((Thetanew,SmoothRadius,X,Y),axis=1)
    SortedCoord = FullCoordinates[FullCoordinates[:,0].argsort(),]
    
    SortedTheta = SortedCoord[:,0]
    SortedRadius = SortedCoord[:,1]
    SortedEdgeX = SortedCoord[:,2]
    SortedEdgeY = SortedCoord[:,3]   
    
    
    return(SortedTheta, SortedRadius, SortedEdgeX, SortedEdgeY)
```

###### Compute PDFs and contours

```python
def PDFmean(StackList,CD,P):
    
    npx = 500 # size of final image
    
    if not os.path.exists(P + '\\Figures\\PDFs\\'):
        os.mkdir(P + '\\Figures\\PDFs\\') # create global folder 
    
    
    if not os.path.exists(P + '\\Figures\\PDFs\\PDF\\'):
        os.mkdir(P + '\\Figures\\PDFs\\PDF\\') # create image folder 
        
    if not os.path.exists(P + '\\Figures\\PDFs\\PDFnorm\\'):
        os.mkdir(P + '\\Figures\\PDFs\\PDFnorm\\') # create image folder 
    
    
    if not os.path.exists(P + '\\Figures\\PDFs\\PDFmean\\'):
        os.mkdir(P + '\\Figures\\PDFs\\PDFmean\\') # create image folder 
        
    if not os.path.exists(P + '\\Figures\\PDFs\\PDFnormmean\\'):
        os.mkdir(P + '\\Figures\\PDFs\\PDFnormmean\\') # create image folder 

    # Number of images in shortest vid
    nmint = 1 + int(np.min([np.max(CD.loc[s, 'Img']) for s in StackList]))

    for ni in range(nmint):
        
        print('Computing PDFs for image : ' + str(ni+1) + '/' + str(nmint) + ' :')
        
        # For PDF storage
        PDFs = np.empty((npx,npx,len(StackList)))
        nPDFs = np.empty((npx,npx,len(StackList)))
    
        for s,i in zip(StackList,range(len(StackList))):
            PPG = CD.loc[s]

            
            print('Computing PDFs for ' + s.ljust(5) , end='\r')
            
            # Non normalized
            
            X = PPG.loc[PPG['Img']==ni,'Xparam']
            Y = PPG.loc[PPG['Img']==ni,'Yparam']

            path = mpltPath.Path([(x,y) for x,y in zip(X,Y)])
            Size = 1000 # in µm
            Scale = Size/npx # in µm/px
            
            xx, yy = np.meshgrid(np.linspace(-Size,Size,npx),np.linspace(-Size,Size,npx))    

            insideBool = path.contains_points([[x,y] for x,y in zip(xx.flatten(),yy.flatten())])

            BoolMat = insideBool.reshape(npx,npx)
            
            PDFs[:,:,i] = BoolMat            
            
            # Normalized

            Xn = PPG.loc[PPG['Img']==ni,'XparamNorm']
            Yn = PPG.loc[PPG['Img']==ni,'YparamNorm']


            pathn = mpltPath.Path([(x,y) for x,y in zip(Xn,Yn)])

            SizeN  = 1.5 # Normalized size
            ScaleN = SizeN/npx # in /px
            
            xxn, yyn = np.meshgrid(np.linspace(-SizeN,SizeN,npx),np.linspace(-SizeN,SizeN,npx))    

            insideBooln = pathn.contains_points([[x,y] for x,y in zip(xxn.flatten(),yyn.flatten())])

            BoolMatn = insideBooln.reshape(npx,npx)

            nPDFs[:,:,i] = BoolMatn

            
        
        print('All PPGs done.'.ljust(25))
            
        fig0, ax0 = plt.subplots(dpi=250)
        ax0.imshow(np.mean(PDFs,2),cmap='jet')
        fig0.savefig(P + '\\Figures\\PDFs\\PDF\\' + str(ni) + '.png') 
        plt.close(fig0)

        fig1, ax1 = plt.subplots(dpi=250)
        ax1.imshow(np.mean(nPDFs,2),cmap='jet')
        fig1.savefig(P + '\\Figures\\PDFs\\PDFnorm\\' + str(ni) + '.png') 
        plt.close(fig1)
        
        # Binary images from PDFs to obtain contours
        PDFbin = np.uint8(np.mean(PDFs,2)>0.5)
        nPDFbin = np.uint8(np.mean(nPDFs,2)>0.5)
        
        fig01, ax01 = plt.subplots(dpi=250)
        ax01.imshow(PDFbin,cmap='gray')
        fig01.savefig(P + '\\Figures\\PDFs\\PDFmean\\' + str(ni) + '.png') 
        plt.close(fig01)

        fig11, ax11 = plt.subplots(dpi=250)
        ax11.imshow(nPDFbin,cmap='gray')
        fig11.savefig(P + '\\Figures\\PDFs\\PDFnormmean\\' + str(ni) + '.png') 
        plt.close(fig11)
        
        Theta, Radius, EdgeX, EdgeY = ContourFromBinImg(PDFbin,Scale)
        
        # Storing contour data
        data = {'Xparam':EdgeX,
                    'Yparam':EdgeY,  
                    'R':Radius,  
                    'Theta':Theta,
                    'Img':ni*np.ones(len(EdgeX))} 
      
        newdata = pd.DataFrame(data=data,index = np.repeat('meanContPDF',len(EdgeX)))
        
        CD = CD.append(newdata) # adding to global dataframe
        
        
        Theta, RadiusN, EdgeXN, EdgeYN = ContourFromBinImg(nPDFbin,ScaleN)
         # Storing contour data
        data = {'Xparam':EdgeXN,
                    'Yparam':EdgeYN,  
                    'R':RadiusN,  
                    'Theta':Theta,
                    'Img':ni*np.ones(len(EdgeXN))} 
      
        newdata = pd.DataFrame(data=data,index = np.repeat('normMeanContPDF',len(EdgeXN)))
        
        CD = CD.append(newdata) # adding to global dataframe
        
    return(CD)
    
print('Function defined')
```

#### In area

Computation of mean contour with different contour grouped based on area rather than timepoint. When several timepoints of the same gemmae are in the same area group they are first averaged together before being averaged with the rest of the gemmae.

```python
def computeMeanContourArea(StackList,CD,GD):
    
    # Number of area groups
    nmax = 1 + int(np.max([np.max(CD.loc[s, 'AreaGroup']) for s in StackList]))
        
    npts = len(CD[CD['Img']==0].loc[StackList[0],'Xparam'])
    
    # Storing variables
    alignedXs = np.empty((npts,len(StackList)))
    alignedXs[:] = np.nan
    alignedYs = np.empty((npts,len(StackList)))
    alignedYs[:] = np.nan
    
    
    for i in range(1,nmax+1):
        
        print('Computing mean contour for area group : ' + str(i) + '/' + str(nmax).ljust(5), end='\r')
        
        # On each image, retrieve aligned contour for each gemma
        
        for ii,s in zip(range(len(StackList)),StackList):
           
            imglist = np.unique(CD.loc[(CD.index == s) & (CD['AreaGroup'] == i),'Img'].to_numpy())

            if not imglist.size==0:
                
                AreaXs = np.empty((npts,len(imglist)))
                AreaXs[:] = np.nan
                AreaYs = np.empty((npts,len(imglist)))
                AreaYs[:] = np.nan
            
                for iii,img in zip(range(len(imglist)),imglist): 
                    
                    Xn = CD.loc[(CD.index == s) & (CD['Img'] == img),'Xparam']
                    Yn = CD.loc[(CD.index == s) & (CD['Img'] == img),'Yparam']

                    AreaXs[:,iii] = Xn
                    AreaYs[:,iii] = Yn
                
                alignedXs[:,ii] = np.nanmean(AreaXs,1)
                alignedYs[:,ii] = np.nanmean(AreaYs,1)

        # Computing mean of aligned contours        
        meanX = np.nanmean(alignedXs,1)      
        meanY = np.nanmean(alignedYs,1) 
      
        
        # Storing contour data
        data = {'Xparam':meanX,
                    'Yparam':meanY, 
                    'AreaGroup':i*np.ones(len(meanX))} 
      
        newdata = pd.DataFrame(data=data,index = np.repeat('meanContArea',len(meanX)))
        
        CD = CD.append(newdata) # adding to global dataframe

    print('Computing mean contour for areagroup : ' + str(nmax) + '/' + str(nmax).ljust(5), end='\n\n')
        
    return(CD)

print('Function defined.')
```

### Computation of growth and similarity 

Similarity : 
Comparison of one normalized contour in time (X(s),Y(s)) with either the first contour of the same gemma - selfSimilarity- or with the average contour -meanSimilarity-.

Growth : 
Percentage of radial extension Gn(theta, t)

```python
def computeGrowthAndSimi(StackList,CD,GD):  
       
    # define storing variables
    CD['Growth'] = np.nan    
    GD['meanSimi'] = np.nan
    GD['selfSimi'] = np.nan 
    GD['meanSimiNorm'] = np.nan    
       
    # Number of image in the shortest and longest videos
    tmax = 1 + int(np.max([np.max(CD.loc[s, 'Img']) for s in StackList]))
    tmin = 1 + int(np.min([np.max(CD.loc[s, 'Img']) for s in StackList]))
      
    npts = len(CD.loc[(CD.index == StackList[0]) & (CD['Img'] == 0),'Xparam'])
    
    GrowthMat = np.empty((len(StackList),npts,tmax)) # Growth Gn(Theta,t) 
        
    for ni,n in zip(range(len(StackList)),StackList):
        
        print('Processing ' + n + ' :')
            
        timg = int(1 + np.max(CD.loc[n, 'Img'])) # number of images
        
        for ti in range(timg):
            
            print('Computing similarities and growth for image : ' + str(ti+1) + '/' + str(timg).ljust(5), end='\r')
            
            # Normalized contour and start (t=0) contour
            Xn = CD.loc[(CD.index == n) & (CD['Img'] == ti),'XparamNorm'].to_numpy()
            Yn = CD.loc[(CD.index == n) & (CD['Img'] == ti),'YparamNorm'].to_numpy()
            Xn0 = CD.loc[(CD.index == n) & (CD['Img'] == 0),'XparamNorm'].to_numpy()
            Yn0 = CD.loc[(CD.index == n) & (CD['Img'] == 0),'YparamNorm'].to_numpy()
            meanXn = CD.loc[(CD.index == 'normMeanContTime') & (CD['Img'] == ti), 'Xparam'].to_numpy()
            meanYn = CD.loc[(CD.index == 'normMeanContTime') & (CD['Img'] == ti), 'Yparam'].to_numpy()
            
            # Computing self similarity
            selfSimi = vf.HausdorffDist(Xn,Yn,Xn0,Yn0)
            GD.loc[(GD.index == n) & (GD['Img'] == ti),'selfSimi'] = selfSimi
            
            # Contour, start contour, and mean contour
            X = CD.loc[(CD.index == n) & (CD['Img'] == ti),'Xparam'].to_numpy()
            Y = CD.loc[(CD.index == n) & (CD['Img'] == ti),'Yparam'].to_numpy()
            X0 = CD.loc[(CD.index == n) & (CD['Img'] == 0),'Xparam'].to_numpy()
            Y0 = CD.loc[(CD.index == n) & (CD['Img'] == 0),'Yparam'].to_numpy()
            meanX = CD.loc[(CD.index == 'meanContTime') & (CD['Img'] == ti), 'Xparam'].to_numpy()
            meanY = CD.loc[(CD.index == 'meanContTime') & (CD['Img'] == ti), 'Yparam'].to_numpy()

            if not ti >= tmin:
                
                # Computing mean similarity
                
                meanSimi = vf.HausdorffDist(X,Y,meanX,meanY)
                meanSimiN = vf.HausdorffDist(Xn,Yn,meanXn,meanYn)

                GD.loc[(GD.index == n) & (GD['Img'] == ti),'meanSimi'] = meanSimi
                GD.loc[(GD.index == n) & (GD['Img'] == ti),'meanSimiNorm'] = meanSimiN
            
                
            # Growth
            #Gn = (np.divide(rn,rn0)-1)*100
            
            #CD.loc[(CD.index == n) & (CD['Img'] == ti),'Growth'] = Gn
            
            #GrowthMat[ni,:,ti] = Gn
        
        print('Computing similarities and growth for image : ' + str(ti+1) + '/' + str(timg).ljust(5), end='\n\n')    
            
    return(CD,GD,GrowthMat)  

print('Function defined.')       
            
```

### Time correlation in growth variations
Comparison of area variation ratio for three time points. Computation of the correlation coefficient for all PPGS. Data stored in GlobalData dataframes, at Img 1 position for each gemma.

```python
def GrowthTimeCorr(StackList, GD, Tpts):
    
    # Tpts are the three time points between which growth correlation is computed, in hours
    
    GD['TimesPoints'] = np.nan
    GD['G01'] = np.nan 
    GD['G12'] = np.nan 
    
    # Mean area for each time point
    MeanA = np.empty(max(GD['Img']+1))
    for im in range(0,max(GD['Img']+1)):
                    
        MeanA[im] = GD[GD['Img'] == im]['Area'].to_numpy()[np.array([GD['GoodFocus']==True])[0][GD['Img'] == im]].mean()
    
    # Time points between which to compute growth correlation ( converted to minutes)
    t0, t1, t2 = np.array(Tpts)*60
    
    for s in StackList:        
                   
        Time = GD.at[s, 'Time (min)'].to_numpy()[GD.at[s, 'GoodFocus']==True]
        Area = GD.at[s, 'Area'].to_numpy()[GD.at[s, 'GoodFocus']==True]

        idxT0 = np.argmin(abs(Time-t0))
        idxT1 = np.argmin(abs(Time-t1))
        idxT2 = np.argmin(abs(Time-t2))

        A0 = Area[idxT0] #-MeanA[idxT0]
        A1 = Area[idxT1] #-MeanA[idxT1]
        A2 = Area[idxT2] #-MeanA[idxT2]
        
        G01 = A1/A0 - 1
        G12 = A2/A1 - 1
        
        GD.loc[(GD.index == s) & (GD['Img'] == 1),'CorrT0'] = t0
        GD.loc[(GD.index == s) & (GD['Img'] == 1),'CorrT1'] = t1
        GD.loc[(GD.index == s) & (GD['Img'] == 1),'CorrT2'] = t2
        GD.loc[(GD.index == s) & (GD['Img'] == 1),'G01'] = G01
        GD.loc[(GD.index == s) & (GD['Img'] == 1),'G12'] = G12
    
    PearsonCorrCoeff = GD[['G01','G12']].corr(method='pearson').at['G01','G12']
    TauKendall = GD[['G01','G12']].corr(method='kendall').at['G01','G12']
    RhoSpearman = GD[['G01','G12']].corr(method='spearman').at['G01','G12']
    
    CorrCoeffs = np.round(np.array([PearsonCorrCoeff, TauKendall, RhoSpearman])*100)/100
    
    return(GD,CorrCoeffs)
```

### Estimating symmetry of the shape

Computes the assymmetry of a contour, with the Y axis as axis of symmetry 

The computation is done by computing the Hausdorff distance between the original contour and the same contour flipped around the symetry axis.


```python
def computeSym(StackList,CD,GD, **kwargs):    
      
    DebugPlots = False
    
    for key, value in kwargs.items(): 
        if key == 'debug':
            DebugPlots = value
        else:
            print('Unknown key : ' + key+ '. Kwarg ignored.')
        
    GD['Assymmetry'] = np.nan
    
    for n in StackList:
        
        timg = int(1 + np.max(CD.loc[n, 'Img'])) # number of images
        
        for ti in range(timg):
            
            X = CD.loc[(CD.index == n) & (CD['Img'] == ti),'Xparam'].to_numpy()
            Y = CD.loc[(CD.index == n) & (CD['Img'] == ti),'Yparam'].to_numpy()
            
            A,R = vf.ToCirc(X,Y,angle='deg')
            
            AS = vf.HausdorffDist(X,Y,-X,Y, debug =DebugPlots) 
            
            
            GD.loc[(GD.index == n) & (GD['Img'] == ti),'Assymmetry'] = AS/np.median(R) # normalization by median radius
            
    return(CD,GD)

print('Function defined.')
```

### Plot gemmae area in time

```python

def plotArea(StackList, GD, P, CCs, Title, **kwargs):

    # Colors
    colorcycle = [plt.get_cmap('gist_rainbow')(1. * i/30) for i in range(30)]
    mpl.rcParams['axes.prop_cycle'] = cycler(color=colorcycle)
    
    #init and read kwargs
    Trap = False # default value
    soloGraphs = False
    
    for key, value in kwargs.items(): 
        if key == 'Trapped':
            Trap = value
        elif key == 'solo':
            soloGraphs = value
        else:
            print('Unknown key : ' + key+ '. Kwarg ignored.')
         
    # check existence of figure folder, if absent, create it
    if not os.path.exists(P + '\\Figures'):
            os.mkdir(P + '\\Figures') # create folder
    
    # plot figure
    
    ###################### plot individual graphs #######################
    
    if soloGraphs: 
    
        for s in StackList:
            

            fig1 = plt.figure(dpi = 200,facecolor='black')
            fig1.suptitle(s + ' - Size vs. time')
            plt.xlabel('Time (min)')
            plt.ylabel('Size (mm²)')
            

            xTime = GD.at[s, 'Time (min)'].to_numpy()[GD.at[s, 'GoodFocus']==True]
        
            
            PropBBOX = GD.at[s, 'Bounding box (px)']

            plt.plot(xTime,GD.at[s, 'Size (mm²)'].to_numpy()[GD.at[s, 'GoodFocus']==True],'-o',ms=3,figure=fig1,label=s)

            # if there is a trap, find out when th ppgls touches up and bottom wall
            if Trap:
                trapdata =vf.LoadImageJResults(P + '\\' + s + '_Trap.txt', ['BX','BY','Width','Height'])
                TrapBBOX = [trapdata.at[0, 'BY'], trapdata.at[0, 'BX'],trapdata.at[0, 'BY'] + trapdata.at[0, 'Height'] ,
                        trapdata.at[0, 'BX'] + trapdata.at[0, 'Width']]

                BBOXDiff = PropBBOX - TrapBBOX            
                idxTouch = np.argwhere(np.logical_and(BBOXDiff[0:,0]<0, BBOXDiff[0:,2]>0))

                plt.plot(xTime[idxTouch],GD.at[s, 'Size (mm²)'][idxTouch],'-or',lw=0.5,ms=1.5,figure=fig1,label='Touching UpDown')
                plt.legend(prop={'size': 4})

                fig1.savefig(P + '\\Figures\\' + s + '_SizeTime.png')    


    ###################### plot common graphs #######################
    
    fig2,ax2 = plt.subplots(dpi = 200,facecolor='black')
    fig2.suptitle(Title + ' - Area vs. time')
    plt.xlabel('Time (min)')
    plt.ylabel('Area (mm²)')
    
    fig3,ax3 = plt.subplots(dpi = 200,facecolor='black')
    fig3.suptitle(Title + ' - Norm Area vs. time')
    plt.xlabel('Time (min)')
    plt.ylabel('Area (normalized)')
    plt.legend(prop={'size': 4})
    
    # Computing mean area over all gemmae for ach image
    MeanA = np.empty(max(GD['Img']+1))
    MeanTime = np.empty(max(GD['Img']+1))
    
    for im in range(0,max(GD['Img']+1)):
                    
        MeanA[im] = GD[GD['Img'] == im]['Area'].to_numpy()[np.array([GD['GoodFocus']==True])[0][GD['Img'] == im]].mean()
        MeanTime[im] = im*30
    
    for s in StackList:        
                   
        xTime = GD.at[s, 'Time (min)'].to_numpy()[GD.at[s, 'GoodFocus']==True]
        Area = GD.at[s, 'Area'].to_numpy()[GD.at[s, 'GoodFocus']==True]
        
        NormArea = Area/Area[0]
        
        ax2.plot(xTime,Area/1000000,'-o',ms=3,label=s)
        ax3.plot(xTime,NormArea,'-o',ms=3,label=s)
        
    
    ax2.plot(MeanTime,MeanA/1000000,'-*w',ms=2,label='Mean')
    ax3.plot(MeanTime,MeanA/MeanA[0],'-*w',ms=2,label='Mean')
    
    plt.figure(fig2.number)
    plt.legend(prop={'size': 4})
    fig2.savefig(P + '\\Figures\\AreaTime_' + Title + '.png')
    
    plt.figure(fig3.number)
    plt.legend(prop={'size': 4})
    fig3.savefig(P + '\\Figures\\NormAreaTime_' + Title + '.png')
    
    
    ###################### plot time correlation #######################
    
    t0 = GD.loc[(GD.index == StackList[0]) & (GD['Img'] == 1),'CorrT0'].to_numpy()
    t1 = GD.loc[(GD.index == StackList[0]) & (GD['Img'] == 1),'CorrT1'].to_numpy()
    t2 = GD.loc[(GD.index == StackList[0]) & (GD['Img'] == 1),'CorrT2'].to_numpy()
    
    fig4,ax4 = plt.subplots(dpi = 200,facecolor='black')
    fig4.suptitle(Title + ' - Area growth time correlation (T = ' + str(int(t0[0])) + ', ' + str(int(t1[0])) + ', ' + str(int(t2[0])) +  ')' + 
                  '\nPCC = ' + str(CCs[0]) + ' - KT = ' + str(CCs[1]) + ' - SR = ' + str(CCs[2]))
    plt.xlabel('G01')
    plt.ylabel('G12')
    plt.legend(prop={'size': 4})
    
    ax4.scatter(GD['G01'],GD['G12'])
    
    fig4.savefig(P + '\\Figures\\GrowthTimeCorr_' + Title + '.png')
    plt.show()
    return    

print('Function defined')
```

#### Plot similarities in time

```python
def plotShapeSimi(P,StackList,CD,GD, figname):
    
    # Number of image in the shortest and longest videos
    nmin = 1 + int(np.min([np.max(CD.loc[s, 'Img']) for s in StackList]))
    nmax = 1 + int(np.max([np.max(CD.loc[s, 'Img']) for s in StackList]))
    
    ### Area and selfSimilarity comparison in time
   
    # initialize figure
    fig0, ax0 = plt.subplots(dpi = 250,facecolor='black')
    fig0.suptitle('selfSimi vs. Area - ' + figname)
    plt.xlabel('Area (norm)')
    plt.ylabel('Similarity to self')                
    
    print('Plot self similarity graphs...', end='')
    
    for s in StackList:
        
        xTime = GD.at[s, 'Time (min)'].to_numpy()[GD.at[s, 'GoodFocus']==True]/60
        Area = GD.at[s, 'Area'].to_numpy()[GD.at[s, 'GoodFocus']==True]        
        NormArea = Area/Area[0]        
        selfSimi = GD.at[s, 'selfSimi'].to_numpy()[GD.at[s, 'GoodFocus']==True]

        ax0.plot(NormArea,selfSimi,'c-', lw = 0.5,zorder=10)
        sc = ax0.scatter(NormArea,selfSimi, c= xTime, cmap = 'jet', vmin=0, vmax = 0.5*nmax,zorder=5)
    
    fig0.colorbar(sc, ax = ax0, label = 'Time (h)')
    
    fig0.savefig(P + '\\Figures\\AreaSelfSimi.png')
    
    print('Done and saved.')
    

    ### Area and meanSimilarity comparison in time
    
    # initialize figure
    fig1, ax1 = plt.subplots(dpi = 250,facecolor='black')
    fig1.suptitle('meanSimi vs. Area - ' + figname)
    plt.xlabel('Area (norm)')
    plt.ylabel('Similarity to mean')  
    
    fig2, ax2 = plt.subplots(dpi = 250,facecolor='black')
    fig2.suptitle('meanSimiNorm vs. Area - ' + figname)
    plt.xlabel('Area (norm)')
    plt.ylabel('Similarity to mean with normalized contours')  
    
    print('Plot mean similarity graphs...', end='')
    
    for s in StackList:
        
        xTime = GD.at[s, 'Time (min)'].to_numpy()[GD.at[s, 'GoodFocus']==True]/60
        Area = GD.at[s, 'Area'].to_numpy()[GD.at[s, 'GoodFocus']==True]        
        NormArea = Area/Area[0]        
        meanSimi = GD.at[s, 'meanSimi'].to_numpy()[GD.at[s, 'GoodFocus']==True]
        meanSimiN = GD.at[s, 'meanSimiNorm'].to_numpy()[GD.at[s, 'GoodFocus']==True]

        
        ax1.plot(NormArea,meanSimi,'c-', lw = 0.5,zorder=10)
        ax2.plot(NormArea,meanSimiN,'c-', lw = 0.5,zorder=10)
        sc1 = ax1.scatter(NormArea,meanSimi, c= xTime, cmap = 'jet', vmin=0, vmax = 0.5*nmax,zorder=5)
        sc2 = ax2.scatter(NormArea,meanSimiN, c= xTime, cmap = 'jet', vmin=0, vmax = 0.5*nmax,zorder=5)
    
    fig1.colorbar(sc1, ax = ax1, label = 'Time (h)')
    fig2.colorbar(sc2, ax = ax2, label = 'Time (h)')
    
    fig1.savefig(P + '\\Figures\\AreaMeanSimi.png')
    fig2.savefig(P + '\\Figures\\AreaMeanSimiNorm.png')
    
    print('Done and saved.')
    plt.show()
    return

print('Function defined.')
```

#### Plot Growth quantifications

Growth Gn(Theta,t) = G(n,Th,t)

```python
def plotGrowth(P,StackList,CD,GD,GrowthMat):
    
    ### Individual growth polar plot    
    
    # Specifics of circular histogram
    N = 20 # number of quadrant
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False) # Abcisse
    width = (2*np.pi) / N # bar width    
    bottom = 0 # Bottom level
    
    if not os.path.exists(P + '\\Figures\\Growth\\'):
        os.mkdir(P + '\\Figures\\Growth\\') # create image folder   
    
    A = CD[CD['Img']==0].loc[StackList[0],'Theta'].values+180
    
    for s in StackList:    
        
        print('Plot growth for gemma : ' + s + '...', end='')
        
        if not os.path.exists(P + '\\Figures\\Growth\\' + s  + '_Growth\\'):
            os.mkdir(P + '\\Figures\\Growth\\' + s  + '_Growth\\') # create save folder
       
        n = int(1 + np.max(CD.loc[s, 'Img']))
        
        for i in range(n):
            
            
            if (GD.loc[s]['GoodFocus'].to_numpy()[i] == True):
                
 
                G = CD[CD['Img']==i].loc[s,'Growth']

                # initialize figure
                fig1 = plt.figure(dpi = 250,facecolor='black')
                fig1.suptitle(s + ' growth - T = ' +str(GD[GD['Img']==i].loc[s,'Time (min)']))
                ax1 = plt.subplot(111, polar=True)
                ax1.set_ylim(ymin = -1.5, ymax = 100)
                ax1.set_yticks([0,25,50,75,100])
                ax1.set_yticklabels(['','','50%','','100%'])
                
                radii = np.empty(len(theta))
                # Compute hist values
                radii[1:] = [G[(A > np.mod(np.rad2deg(th-width/2),360)) & (A < np.mod(np.rad2deg(th+width/2),360))].mean() for th in theta[1:]]
                radii[0] = G[(A > np.mod(np.rad2deg(theta[0]-width/2),360)) | (A < np.mod(np.rad2deg(theta[0]+width/2),360))].mean()
                
                bars = ax1.bar(theta, radii, width=width, bottom=bottom)

                # Use custom colors
                for r, bar in zip(radii, bars):
                    bar.set_facecolor(plt.cm.jet(r / 100.))
                
                fig1.savefig(P + '\\Figures\\Growth\\' + s  + '_Growth\\' + str(i) + '.png') 
                
                plt.close(fig1.number)
                
                del G
                
        print('Done and saved.\n')
    
    print('Plotting average growth...')

    if not os.path.exists(P + '\\Figures\\Growth\\' + 'MeanGrowth\\'):
            os.mkdir(P + '\\Figures\\Growth\\' + 'MeanGrowth\\') # create save folder

    nmax = 1 + int(np.max([np.max(CD.loc[s, 'Img']) for s in StackList]))
   

    for i in range(nmax):
        # initialize figure
        fig1 = plt.figure(dpi = 250,facecolor='black')
        fig1.suptitle('Meangrowth - T = ' + str(GD[GD['Img']==i].loc[StackList[0],'Time (min)']))
        ax1 = plt.subplot(polar=True)
        ax1.set_ylim(ymin = -1.5, ymax = 100)
        ax1.set_yticks([0,25,50,75,100])
        ax1.set_yticklabels(['','','50%','','100%'])

        
        radii = np.empty(len(theta))
        # Compute hist values
        radii[1:] = [np.nanmean(GrowthMat[:,(A > np.mod(np.rad2deg(th-width/2),360)) &
                           (A < np.mod(np.rad2deg(th+width/2),360)),i]) for th in theta[1:]]
        
        radii[0] = GrowthMat[:,(A > np.mod(np.rad2deg(theta[0]-width/2),360)) |
                             (A < np.mod(np.rad2deg(theta[0]+width/2),360)),i].mean()
        
        bars = ax1.bar(theta, radii, width=width, bottom=bottom)
        
        # Use custom colors
        for r, bar in zip(radii, bars):
            bar.set_facecolor(plt.cm.jet(r / 100.))

        fig1.savefig(P + '\\Figures\\Growth\\' + 'MeanGrowth\\' + str(i) + '.png') 

        plt.close(fig1.number)
    plt.show()
    return

print('Function defined.')
    
```

```python
def plotAssymetry(P, StackList,GD):
    
    colorcycle = [plt.get_cmap('gist_rainbow')(1. * i/len(StackList)) for i in range(len(StackList))]
    mpl.rcParams['axes.prop_cycle'] = cycler(color=colorcycle)
    
   # Assymmetry in time
    fig0, ax0 = plt.subplots(dpi = 250,facecolor='black')
    fig0.suptitle('Assymmetry in time')     

    # Assymmetry vs area
    fig1, ax1 = plt.subplots(dpi = 250,facecolor='black')
    fig1.suptitle('Assymmetry vs area')   
    
    for s in StackList:
        
        xTime = GD.loc[s,'Time (min)']
        Assym = GD.loc[s,'Assymmetry']
        Area = GD.loc[s,'Area']
            
        ax0.plot(xTime,Assym,'-o',ms = 3)
        
        ax1.plot(Area,Assym,'-o', ms = 3)
        
    fig0.legend(StackList,prop={'size': 5})
    fig1.legend(StackList,prop={'size': 5})
    
    fig0.savefig(P + '\\Figures\\AssymTime.png') 
    fig1.savefig(P + '\\Figures\\AssymArea.png') 
    plt.show()
    return

print('Function defined.')

    
```
