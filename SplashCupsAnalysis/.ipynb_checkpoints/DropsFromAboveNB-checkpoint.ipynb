# 2. Loading stack and locating objects

Here '21' is the estimate size of particles to track.
Minmass the minimal mass (integrated pixel value) for a particule, it is really high because image are 16bits, it may be easier to convert in 8bits before

```python
def LoadAndLocate(filename):
    stack = pims.open(filename) # get the tiff stack
    tp.quiet() # turn off display
    features = tp.batch(stack, 21, minmass=100000, invert=True); # equivalent to tp.locate but on all frames specified
    return(stack, features)
```
